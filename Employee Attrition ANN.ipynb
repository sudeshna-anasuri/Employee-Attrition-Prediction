```python
from sklearn.model_selection import train_test_split                #previously cros_validation was used in sklearn
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler#calculates mean and standard deviation,it gets alldata into same range.(mean)
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
import keras #Building neural network model back end it has tensorflow
from keras.models import Sequential # initializing model
from keras.layers import Dense # used for adding the neural network layers
```

```python
model=Sequential()#instance
```

```python
model.add(Dense(input_dim=22,init="random_uniform",activation="relu",output_dim=18)) #input_dim=no of neurons to have in the input layer(no of independent variables),init=adding weights,activation function,output_dim=next hidden layer
#best thing for output_dim is input_dim/2
```

```python
model.add(Dense(output_dim=14,init="random_uniform",activation="relu"))# adding hidden layer similary like adding weights
```

```python
model.add(Dense(input_dim=1,init="random_uniform",activation='sigmoid',output_dim=1))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])#adam=batch gradent descent
```

```python
model.fit(x_train,y_train,epochs=1000,batch_size=32)#epochs no of iteration 
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
y_pred=(y_pred>0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
